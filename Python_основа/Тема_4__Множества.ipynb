## **Множества (`set`)**

1. **Множество** – это неупорядоченный и неиндексируемый набор данных. Т.е. к его элементам не получить доступа по индексу.
2. Множество хранит только уникальные объекты. То есть во множестве не может повторяться один и тот же элемент.
3. Множества с какими-либо элементами заключают в фигурные `{}` скобки. Элементы идут через запятую (никаких двоеточий, в отличие от словаря).
4. Важно: для создания пустого множества используют только функцию `set()`.

```python
# Создание пустого множества. Только один вариант!

empty_set = set()
print(empty_set, '  ', type(empty_set))
```

Не получится создать пустое множество с помощью фигурных скобок, так как этот способ уже зарезервирован для словарей:

```python
# Так создавать пустое множество НЕЛЬЗЯ!

empty_set2 = {}
type(empty_set2)
```

Но можно и создавать множества с элементами с помощью фигурных скобок, перечисляя элементы через запятую:

```python
# Создание множества с элементами
my_set = {1, 3.5, 'котик', -123, 1, 3.5, 'котик', -123,1, 3.5, 'котик', -123,1, 3.5, 'котик', -123,1, 3.5, 'котик', -123}
print(my_set, '  ', type(my_set))
```

Теперь попробуйте взять какой-то элемент этого множества, используя индекс:

```python
# Попытка обратиться к элементу множества
my_set[1]
```

Как видите, Python выдал ошибку. Это происходит потому, что множества не поддерживают индексацию элементов.

Однако по множествам можно итерироваться с помощью цикла `for`:

```python
# По множеству можно итерироваться циклом for
for element in my_set:
  print(element)
```

Обратите внимание, что элементы были выведены на экран в произвольном порядке, то есть отличном от того, который вы указывали при создании. Это одна из особенностей множества: элементы хранятся в произвольном порядке.

Еще одна особенность: сколько бы повторяющихся элементов вы не добавляли, множество будет хранить лишь уникальные значения. То есть в множестве невозможно встретить дубликаты (одинаковые объекты).

```python
# Сколько повторяющихся элементов не подавать во множество, оно сохранит только уникальные, стирая дубликаты
my_set = {1, 3.5, 'котик', -123}
my_set2 = {1, 3.5, 'котик', -123, 'котик', 'котик', 'котик'}
print(my_set2)
print('Равны ли 1-е и 2-е множества?', my_set == my_set2)
```

Вы можете видеть, как все четыре котика превращаются в одного и множество `my_set` оказывается равным множеству `my_set2`.

Ради эксперимента попробуйте сделать то же самое со списками. Создайте два списка с одинаковыми элементами, но разным количеством. Проверьте, равны ли эти списки между собой:

```python
# Проверьте равенство списков с дубликатами

new_list1 = [1, 3.5, 'котик', -123]
new_list2 = [1, 3.5, 'котик', -123, 'котик', 'котик', 'котик']

print('Равны ли 1е и 2е списки?', new_list1 == new_list2)
```

Как видите, списки не одинаковые. В отличие от множеств, они могут хранить в себе сколько угодно повторяющихся элементов.

### **Методы множеств**

Метод `discard` удаляет указанный элемент из множества, если он там есть:

```python
shop_set = {'хлеб', 'молоко', 'мясо', 'творог', 'ягоды'}
print('Множество до:', shop_set)
```

```python
shop_set.discard('молоко') # безопасное удаление
print('Множество после:', shop_set)
```

Метод `clear` удаляет из множества все элементы:

```python
print('Множество до:', shop_set)
shop_set.clear()
```

```python
print('Множество после:', shop_set)
```

Примеры разности множеств `difference()` и пересечения множеств
`intersection()`


```python
shop_set_a = {'хлеб', 'молоко', 'мясо', 'торт', 'творог', 'ягоды'}
shop_set_b = {'хлеб', 'арбуз', 'молоко', 'мясо', 'творог', 'ягоды'}
shop_set_c = {'222','хлеб','1', 'торт', 3, 4, 5, 'ягоды'}

# Разности множеств
# из покупок `shop_set_a` убрали покупки `shop_set_b`
print(shop_set_a.difference(shop_set_b))

# из покупок `shop_set_a` убрали покупки `shop_set_c` и `shop_set_b`
print(shop_set_a.difference(shop_set_c, shop_set_b)) # Пустое множество

# из покупок `shop_set_b` убрали покупки `shop_set_a`
print(shop_set_b.difference(shop_set_a))
print('-------------')

# Пересечение множеств
print(shop_set_a.intersection(shop_set_b))
print("Пересечение трёх множеств:", shop_set_a.intersection(shop_set_c, shop_set_b))
```

**Методы множеств**

- `add()`	добавляет элемент во множество;

- `clear()`	удаляет все элементы из множества;

- `copy()` возвращает копию множества;

- `difference()`	возвращает разницу двух или более множеств в качестве нового множества;

- `discard()`	удаляет элемент из множества, если он в нем содержится;

- `intersection()`	возвращает пересечение множеств в качестве нового множества;

- `pop()`	удаляет и возвращает произвольный элемент в множество. Выдает `KeyError`,  если множество пусто;

- `remove()`	удаляет элемент из набора. Если элемент не является членом множества, выдает `KeyError`;

- `union()`	возвращает объединение множеств в новом наборе;

- `update()` обновляет множество объединением с другим множеством.

```python
a = set()  # пустое множество

b = {}    # пустой словарь
```
