   `difference()`   
`intersection()`


```python
shop_set_a = {'', '', '', '', '', ''}
shop_set_b = {'', '', '', '', '', ''}
shop_set_c = {'222','','1', '', 3, 4, 5, ''}

#  
#   `shop_set_a`   `shop_set_b`
print(shop_set_a.difference(shop_set_b))

#   `shop_set_a`   `shop_set_c`  `shop_set_b`
print(shop_set_a.difference(shop_set_c, shop_set_b)) #  

#   `shop_set_b`   `shop_set_a`
print(shop_set_b.difference(shop_set_a))
print('-------------')

#  
print(shop_set_a.intersection(shop_set_b))
print("  :", shop_set_a.intersection(shop_set_c, shop_set_b))
```
